# Droptest 1

On Friday May 16th we performed a partial height drop test of the avionics stack using the [Dryden Droptower at PSU](http://www.ddt.pdx.edu/). This allowed us to test the high-g performance of the stack by taking advantage of the high braking forces of the drop tower payload carrier at the bottom of the drop. This was the second drop tried in the day.

## Zoom to drop

Lets zoom in to the interesting part.

```python
start = 161.3
end = 163.8
release = 161.471
touch = 162.641
brakes = 163.1

ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x, 'r', lw=0.8, alpha=1, label="x acc")
plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Drop Overview')
ax.axes[0].set_xlim([start, end])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.annotate('Release', xy=(release, 10.2), xycoords='data', xytext=(50, 110), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.annotate('Zero g', xy=(162.3, 5),  xycoords='data', xytext=(0, 0), textcoords='offset points')

plt.annotate('Hit brakes', xy=(brakes, 1), xycoords='data', xytext=(-120, 150), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.show()
```

Aside from the big ringing, if we look at the zero g part we do see some interesting wobbles near the begining of the drop.
