# Droptest 0

On Friday May 16th we performed a full height drop test of the avionics stack using the [Dryden Droptower at PSU](http://www.ddt.pdx.edu/). This allowed us to test the high-g performance of the stack by taking advantage of the high braking forces of the drop tower payload carrier at the bottom of the drop.

## First glance at drop data

Lets take a glance at the file to figure out when we dropped and the aproximate loading.

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 14})

g_0 = 9.80665

columns = numpy.loadtxt("ADIS.csv", delimiter=',', comments='#', unpack=True)

seqn = columns[0]
timestamp = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]

# Get time in seconds from nanoseconds
adis_time = numpy.divide(timestamp, 1e9)

ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x, 'r', lw=0.8, alpha=0.8, label="x acc")
plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Drop Overview')
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.show()
```

Well, we see the short dip to zero, and a huge (18?) g spike when it hits the brakes at the bottom.

## Zoom to drop

Lets zoom in to the interesting part.

```python
start = 160.5
end = 163.6
release = 160.757
touch = 162.641
brakes = 162.867

ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x, 'r', lw=0.8, alpha=1, label="x acc")
plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Drop Overview')
ax.axes[0].set_xlim([start, end])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.annotate('Release', xy=(release, 10.2), xycoords='data', xytext=(50, 110), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.annotate('Ringing', xy=(160.8, 8), xycoords='data', xytext=(40, 70), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.annotate('Zero g', xy=(161.6, 5),  xycoords='data', xytext=(0, 0), textcoords='offset points')

plt.annotate('Carrier bottoms out', xy=(touch, 1), xycoords='data', xytext=(-200, 110), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.annotate('Hit brakes', xy=(brakes, 1), xycoords='data', xytext=(-120, 150), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.show()
```

```python
# get accel in g's
acc_x_g = numpy.multiply(acc_x, 1/g_0)
ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x_g, 'r', lw=0.8, alpha=1, label="x acc")
plt.xlabel('Time [s]')
plt.ylabel('Accel [g]')
plt.title('Drop Zero g Detail')
ax.axes[0].set_xlim([release, brakes])
ax.axes[0].set_ylim([-0.6, 1.45])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.show()
```

Aside from the big ringing, if we look at the zero g part we do see some interesting wobbles near the begining of the drop.

## Zero g Quality

Switching the units to g and zooming in further:

```python
ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x_g, 'r', lw=0.8, alpha=1, label="x acc")
plt.xlabel('Time [s]')
plt.ylabel('Accel [g]')
plt.title('Drop Zero g Quality')
ax.axes[0].set_xlim([release, brakes])
ax.axes[0].set_ylim([-0.1, 0.1])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.show()
```

```python
ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, gyro_x, 'r', lw=0.6, alpha=0.8, label="x gyro")
plt.plot(adis_time, gyro_y, 'b', lw=0.6, alpha=0.8, label="y gyro")
plt.plot(adis_time, gyro_z, 'g', lw=0.6, alpha=0.8, label="z gyro")
plt.xlabel('Time [s]')
plt.ylabel('Rotation Rage [deg/s]')
plt.title('Drop Zero g Detail -- Gyro')
ax.axes[0].set_xlim([start, end])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.legend(loc=2)
plt.show()
```

```python
ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_y, 'b', lw=0.6, alpha=0.8, label="y gyro")
plt.plot(adis_time, acc_z, 'g', lw=0.6, alpha=0.8, label="z gyro")
plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Drop Zero g Detail -- Y/Z Axis Accel')
ax.axes[0].set_xlim([start, end])
ax.axes[0].set_ylim([-1, 1])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)

plt.legend(loc=4)
plt.show()
```

It's hard to say if the device ever really settles.

# Power Draw During Drop

Back to what we were actually going here to measure. What did the power draw look like and was there any anomalous spikes during the main shock.

```python
columns = numpy.loadtxt("RNHP.csv", delimiter=',', comments='#', unpack=True)

seqn = columns[0]
timestamp = columns[1]
fc = columns[2]
raspi = columns[3]
imu = columns[5]
roll = columns[8]

# Get time in seconds from nanoseconds
rnh_time = numpy.divide(timestamp, 1e9)

ax = plt.figure(figsize=(16,9))
plt.plot(rnh_time, fc, 'r', lw=0.8, alpha=0.8, label="FC")
plt.plot(rnh_time, raspi, 'b', lw=0.8, alpha=0.8, label="RasPi Cam")
plt.plot(rnh_time, imu, 'g', lw=0.8, alpha=0.8, label="IMU")
plt.plot(rnh_time, roll, 'c', lw=0.8, alpha=0.8, label="Roll")
plt.xlabel('Time [s]')
plt.ylabel('Current Draw [A]')
plt.title('Drop Power Usage')
ax.axes[0].set_xlim([start, end])
ax.axes[0].set_ylim([0, 1.4])
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)
plt.legend(loc=1)

plt.show()
```
