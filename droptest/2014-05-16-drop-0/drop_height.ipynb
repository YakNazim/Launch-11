# Guessing The Drop Height

Given that we have high quality accelerometer data, can we guess the height of the droptower?

```python
%matplotlib inline

import numpy
from scipy.integrate import simps
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams.update({'font.size': 14})

g_0 = 9.80665

columns = numpy.loadtxt("ADIS.csv", delimiter=',', comments='#', unpack=True)
seqn = columns[0]
timestamp = columns[1]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]

# Time in seconds
adis_time = numpy.divide(timestamp, 1e9)


# Time markers
start = 160.2
pre_drop = 160.7
zg_start_good = 161.5
zg_end_good = 162.6
end = 163.6
```

```python
# De-bias

begin_idx = (numpy.abs(adis_time-zg_start_good)).argmin()
end_idx = (numpy.abs(adis_time-zg_end_good)).argmin()

adis_time_zg = adis_time[begin_idx:end_idx]
acc_x_zg = acc_x[begin_idx:end_idx]
acc_y_zg = acc_y[begin_idx:end_idx]
acc_z_zg = acc_z[begin_idx:end_idx]

acc_x_zero = numpy.average(acc_x_zg)
acc_y_zero = numpy.average(acc_y_zg)
acc_z_zero = numpy.average(acc_z_zg)

print "X_0:", acc_x_zero
print "Y_0:", acc_y_zero
print "Z_0:", acc_z_zero

# Apply bias
acc_x = numpy.subtract(acc_x, acc_x_zero)
acc_y = numpy.subtract(acc_y, acc_y_zero)
acc_z = numpy.subtract(acc_z, acc_z_zero)
```

```python
# Rotate

begin_idx = (numpy.abs(adis_time-(pre_drop-1))).argmin()
end_idx = (numpy.abs(adis_time-pre_drop)).argmin()

adis_time_pre = adis_time[begin_idx:end_idx]
acc_x_pre = acc_x[begin_idx:end_idx]
acc_y_pre = acc_y[begin_idx:end_idx]
acc_z_pre = acc_z[begin_idx:end_idx]

acc_x_pre_0 = numpy.average(acc_x_pre)
acc_y_pre_0 = numpy.average(acc_y_pre)
acc_z_pre_0 = numpy.average(acc_z_pre)

print "X_0:", acc_x_pre_0
print "Y_0:", acc_y_pre_0
print "Z_0:", acc_z_pre_0

# Make normal
vec = numpy.array([acc_x_pre_0, acc_y_pre_0, acc_z_pre_0])
norm = numpy.linalg.norm(vec)

vec_hat = numpy.divide(vec, norm)
angle = numpy.arccos(numpy.dot(vec_hat, [-1,0,0]))
axis = numpy.cross(vec_hat, [-1,0,0])
axis = numpy.divide(axis, numpy.linalg.norm(axis))

print "Normal:", norm
print "Unit vector:", vec_hat
print "Off Angle [deg]", numpy.degrees(angle)
print "Rotation axis", axis

def rotation_matrix(axis, theta):
    """make a rotation matrix for rotating angle theta along axis
    
    :param axis vector: Normalized rotation axis
    :param theta float: Angle to rotate in radians
    :returns: Rotation matrix
    """
    a = numpy.cos(theta/2)
    b,c,d = -axis*numpy.sin(theta/2)
    return numpy.array([[a*a+b*b-c*c-d*d, 2*(b*c-a*d), 2*(b*d+a*c)],
                     [2*(b*c+a*d), a*a+c*c-b*b-d*d, 2*(c*d-a*b)],
                     [2*(b*d-a*c), 2*(c*d+a*b), a*a+d*d-b*b-c*c]])

rot = rotation_matrix(axis, -angle)

print "Rotation Matrix:\n", rot

# Rotate
for i in xrange(len(adis_time)):
    a = numpy.array([acc_x[i], acc_y[i], acc_z[i]])
    r = numpy.dot(rot,a)
    acc_x[i] = r[0]
    acc_y[i] = r[1]
    acc_z[i] = r[2]
```

```python
ax = plt.figure(figsize=(16,9))
plt.plot(adis_time, acc_x, 'r', lw=0.8, alpha=0.8, label="x acc")
plt.plot(adis_time, acc_y, 'b', lw=0.8, alpha=0.6, label="y acc")
plt.plot(adis_time, acc_z, 'g', lw=0.8, alpha=0.6, label="z acc")

plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Normalized, Rotated Drop')
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)
ax.axes[0].set_xlim([start, end])
ax.axes[0].set_ylim([-1, 1])

plt.show()
```
