```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 14})

g_0 = 9.80665

columns = numpy.loadtxt("ADIS.csv", delimiter=',', comments='#', unpack=True)

seqn = columns[0]
timestamp = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]
mag_x = columns[9]
mag_y = columns[10]
mag_z = columns[11]

# Get time in seconds from nanoseconds
adis_time = numpy.divide(timestamp, 1e9)

zg_start_good = 161.5
zg_end_good = 162.6


# De-bias

begin_idx = (numpy.abs(adis_time-zg_start_good)).argmin()
end_idx = (numpy.abs(adis_time-zg_end_good)).argmin()

adis_time_zg = adis_time[begin_idx:end_idx]
acc_x_zg = acc_x[begin_idx:end_idx]
acc_y_zg = acc_y[begin_idx:end_idx]
acc_z_zg = acc_z[begin_idx:end_idx]

acc_x_zero = numpy.average(acc_x_zg)
acc_y_zero = numpy.average(acc_y_zg)
acc_z_zero = numpy.average(acc_z_zg)

# Apply bias
acc_x = numpy.subtract(acc_x, acc_x_zero)
acc_y = numpy.subtract(acc_y, acc_y_zero)
acc_z = numpy.subtract(acc_z, acc_z_zero)

drop_start = 160.767
drop_stop = 162.641

# Cut out just drop time
begin_idx = (numpy.abs(adis_time-161.0)).argmin()
end_idx = (numpy.abs(adis_time-162.5)).argmin()

adis_time_zg = adis_time[begin_idx:end_idx]
acc_x_zg = acc_x[begin_idx:end_idx]
acc_y_zg = acc_y[begin_idx:end_idx]
acc_z_zg = acc_z[begin_idx:end_idx]

# Put in units of g
acc_x_zg = numpy.divide(acc_x_zg, g_0)
acc_y_zg = numpy.divide(acc_y_zg, g_0)
acc_z_zg = numpy.divide(acc_z_zg, g_0)

```

```python
p = numpy.abs(numpy.fft.rfft(acc_x_zg))
h = numpy.multiply(numpy.fft.rfftfreq(len(acc_x_zg)), 819.2)

# Peak power
n = p.argmax()
freq = h[n]

ax = plt.figure(figsize=(16,9))
plt.plot(h, p, 'b', lw=1.5, alpha=0.8, label="Frequency")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')
plt.title('Drop FFT')
ax.axes[0].grid(color='grey', alpha=0.3, linestyle='dashed', linewidth=0.5)
ax.axes[0].set_xlim([0, 400])

plt.annotate('Peak: %0.2f Hz' % freq, xy=(h[n], p.max()), xycoords='data', xytext=(50, -130), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.show()
```
