```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 14})
```

# Droptest Baseline 0

To start off with we made a baseline measurment by leaving the cart still and looking at the power consumption.

The Accelerometer data looked like this:

```python
columns = numpy.loadtxt("ADIS.csv", delimiter=',', unpack=True)

seqn = columns[0]
timestamp = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]

# Get time in seconds from nanoseconds
timestamp = numpy.divide(timestamp, 1e9)

ax = plt.figure(figsize=(16,9))
plt.plot(timestamp, gyro_x, 'r', alpha=0.4, label="x gryo")
plt.plot(timestamp, gyro_y, 'b', alpha=0.3, label="y gryo")
plt.plot(timestamp, gyro_z, 'g', alpha=0.3, label="z gryo")
plt.xlabel('Time [s]')
plt.ylabel('Rate [deg/s]')
plt.title('Still Cart')
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)
plt.legend(loc=2)

plt.annotate('Unplug umbilical', xy=(5.8, -1),  xycoords='data', xytext=(50, -175), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.annotate('Bump cart', xy=(31.5, -1),  xycoords='data', xytext=(-100, -175), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.annotate('Plug in umbilical', xy=(47, -1),  xycoords='data', xytext=(50, -175), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)


plt.show()
```

# Looking at Battery Pack

Take a quick look at the battery pack

```python
columns = numpy.loadtxt("RNHH.csv", delimiter=',', unpack=True)

bat_time = columns[1]
temp = columns[2]
voltage = columns[6]
current = columns[7]

# Get time in seconds from nanoseconds
bat_time = numpy.divide(bat_time, 1e9)

ax = plt.figure(figsize=(16,9))
plt.plot(bat_time, voltage, 'r', lw=2, alpha=0.8, label="Voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.title('Battery Pack')
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)

plt.annotate('Unplug umbilical', xy=(5.8, 16.63), xycoords='data', xytext=(50, -110), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)

plt.annotate('Plug in umbilical', xy=(47, 16.62),  xycoords='data', xytext=(-200, -110), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.annotate('Discharging', xy=(22, 16.45),  xycoords='data', xytext=(0, 0), textcoords='offset points')


plt.show()
```

And total current:

```python
ax = plt.figure(figsize=(16,9))
plt.plot(bat_time, current, 'b', lw=2, alpha=0.8, label="Current")
plt.xlabel('Time [s]')
plt.ylabel('Current [Amp]')
plt.title('Battery Pack')
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)

plt.annotate('Discharging', xy=(22, -1.2),  xycoords='data', xytext=(0, 0), textcoords='offset points')

plt.show()
```

# Port Power

For this test we had the following layout:

 1. Flight computer (FC)
 2. RasPi Cam (PiCam)
 3. NC
 4. IMU board (IMU)
 5. UMB (ignore)
 6. NC
 7. Roll board (Roll)
 8. NC

```python
columns = numpy.loadtxt("RNHP.csv", delimiter=',', unpack=True)

timestamp = columns[1]
FC = columns[2]
PiCam = columns[3]
IMU = columns[5]
Roll = columns[8]

# Get time in seconds from nanoseconds
timestamp = numpy.divide(timestamp, 1e9)

ax = plt.figure(figsize=(16,9))
plt.plot(timestamp, FC, 'r', alpha=1, label="[1] FC")
plt.plot(timestamp, PiCam, 'b', alpha=1, label="[2] PiCam")
plt.plot(timestamp, IMU, 'g', alpha=1, label="[4] IMU")
plt.plot(timestamp, Roll, 'c', alpha=1, label="[7] Roll")

plt.title('Port Power')
plt.xlabel('Time [s]')
plt.ylabel('Current [Amp]')

ax.axes[0].set_ylim([0, 1.6])
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)
plt.legend(loc=1)

plt.show()
```

There are very interesting features in the RasPi camera that looks like it's tracking the big current swings in the flight computer. This is probably because we a reading the ADC too close together and one measurment is effecting the other.

## Total Current Draw

If we add up all the port currents, it should match the battery current draw numbers from before.

```python
ports = numpy.add(FC, PiCam)
ports = numpy.add(ports, IMU)
ports = numpy.add(ports, Roll)
ports = numpy.multiply(ports, -1)

ax = plt.figure(figsize=(16,9))
plt.plot(bat_time, current, 'b', lw=1.5, alpha=0.7, label="Battery Pack Current")
plt.plot(timestamp, ports, 'r', lw=1.5, alpha=0.8, label="Total Port Current")
plt.xlabel('Time [s]')
plt.ylabel('Current [Amp]')
plt.title('Total Current vs Battery Current')
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)
plt.legend(loc=2)
plt.show()
```
