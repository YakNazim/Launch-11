# Data First Look

Just to start out let's look at the data we recovered from the rocket!

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'font.size': 14})
```

## Launch Time

Let's look at the timestamp from the umbilical disconnect. For convinience we'll call this $t_0$ from now on.

```python
with open('fcf/RNHU.csv', 'r') as rnhu:
    for line in rnhu:
        print line

t_0 = 202945201730391 # nanoseconds in FC time frame
```

## Acceleration

Now let's see if the rocket moved! (It did)

```python
columns = numpy.loadtxt("fcf/ADIS.csv", delimiter=',', unpack=True)

seqn = columns[0]
timestamp = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]

timestamp = numpy.subtract(timestamp, t_0)
timestamp = numpy.divide(timestamp, 1e9)

# Peak Thrust
thrust_peak = numpy.argmax(acc_x[0:25000])
thrust_peak = (timestamp[thrust_peak], acc_x[thrust_peak])

# Peak Decel
decel_peak = numpy.argmin(acc_x[0:25000])
decel_peak = (timestamp[decel_peak], acc_x[decel_peak])

# NSR Fire
nsr_event = numpy.argmax(acc_x[27700:27800])
nsr_event = (timestamp[27700+nsr_event], acc_x[27700+nsr_event])

ax = plt.figure(figsize=(16,9))
plt.plot(timestamp, acc_x, 'r', alpha=0.8, label="UP Measured Acceleration")
plt.xlabel('Time [s]')
plt.ylabel('Accel [m/s/s]')
plt.title('Launch 11 Acceleration')
ax.axes[0].grid(color='grey', alpha=0.2, linestyle='dashed', linewidth=0.5)
ax.axes[0].set_xlim([-5, 300])
ax.axes[0].set_ylim([-50, 150])

plt.annotate('Peak Thrust (%0.1f m/s/s)' % thrust_peak[1], xy=(thrust_peak[0], thrust_peak[1]),  xycoords='data',
    xytext=(50, 20), textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)
plt.annotate('NSR Fire (T+%0.3f)' % nsr_event[0], xy=(nsr_event[0], nsr_event[1]),  xycoords='data', xytext=(25, 160),
    textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)
plt.annotate('Drouge Deploy', xy=(35, 15),  xycoords='data', xytext=(60, 100), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
)
plt.annotate('Main Deploy', xy=(225.1, 10),  xycoords='data', xytext=(-100, 100), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)
plt.annotate('Ground Impact', xy=(279, 10),  xycoords='data', xytext=(-150, 220), textcoords='offset points',
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-.2")
)

plt.show()
```
