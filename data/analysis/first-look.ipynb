# First Look at the Flight Data

Just to start out let's look at the data we recovered from the rocket!

```python
%matplotlib inline
import numpy
import utils
from matplotlib import rcParams
rcParams.update({'font.size': 12})

# constants
g_0 = 9.80665      # standard gravity, g, m/s/s
p_0 = 101.326      # sea level, kPa
```

## Launch Time

Let's look at the timestamp from the umbilical disconnect. For convinience we'll call this $t_0$ from now on.

```python
with open('fcf/RNHU.csv', 'r') as rnhu:
    for line in rnhu:
        print line,

t_0 = 202945201730391 # nanoseconds in FC time frame
```

## Acceleration

Now let's see if the rocket moved! (It did)

```python
columns = numpy.loadtxt("fcf/ADIS.csv", delimiter=',', unpack=True)

seqn = columns[0]
timestamp = columns[1]
gyro_x = columns[3]
gyro_y = columns[4]
gyro_z = columns[5]
acc_x = columns[6]
acc_y = columns[7]
acc_z = columns[8]

timestamp = numpy.subtract(timestamp, t_0)
timestamp = numpy.divide(timestamp, 1e9)

# Peak Thrust
thrust_peak = numpy.argmax(acc_x[0:25000])
thrust_peak = (timestamp[thrust_peak], acc_x[thrust_peak])

# Peak Decel
decel_peak = numpy.argmin(acc_x[0:25000])
decel_peak = (timestamp[decel_peak], acc_x[decel_peak])

# NSR Fire
nsr_event = numpy.argmax(acc_x[27700:27800])
nsr_event = (timestamp[27700+nsr_event], acc_x[27700+nsr_event])

fig = utils.Plot("Launch 11 Measured Acceleration", "Time [min:sec]", "Accel [m/s/s]")
fig.plot(timestamp, acc_x, label="UP Measured Acceleration")
fig.xlim([-5, 300])
fig.ylim([-50, 150])
fig.timex()

fig.note('Peak Thrust (%0.1f m/s/s)' % thrust_peak[1], (thrust_peak[0], thrust_peak[1]), (50,20))
fig.note('NSR Fire (T+%0.3f)' % nsr_event[0], (nsr_event[0], nsr_event[1]), (25, 160))
fig.note('Drouge Deploy', (35, 15), (60, 100))
fig.note('Main Deploy', (225.1, 10), (-100, 100))
fig.note('Ground Impact', (279, 10), (-150, 220))
fig.show()
```

Now lets look directly at the part of the flight we really care about, the climb!

```python
fig = utils.Plot("Launch 11 Measured Acceleration", "Time [sec]", "Accel [m/s/s]")
fig.plot(timestamp[2000:28000], acc_x[2000:28000], label="UP Measured Acceleration")
fig.xlim([-1, 32])
fig.ylim([-40, 110])
fig.show()
```

Let's look at lateral g's as well:

```python
fig = utils.Plot("Launch 11 Measured Acceleration", "Time [sec]", "Accel [m/s/s]")
fig.plot(timestamp[2000:28000], acc_x[2000:28000], color='black', alpha=0.13, label="UP Measured Acceleration")
fig.plot(timestamp[2000:28000], acc_y[2000:28000], color=utils.green, label="'Y' Measured Acceleration")
fig.plot(timestamp[2000:28000], acc_z[2000:28000], color=utils.blue, label="'Z' Measured Acceleration")
fig.xlim([-1, 32])
fig.ylim([-40, 110])
fig.legend()
fig.show()
```

It's surprisingly noisy and not aligned. This might be due to high rotation effects in one direction.

## Compared to Simulation

Now we'll bring in the results from [OpenRocket](http://openrocket.sourceforge.net/) and see if it matches up.

```python
columns = numpy.loadtxt("../../sim/flight.csv", delimiter=',', unpack=True)
sim_time = columns[0]
acc_up = columns[3]

# shift into measured accel
acc_up = numpy.add(acc_up, g_0)

fig = utils.Plot("Launch 11 Simulated Acceleration", "Time [sec]", "Accel [m/s/s]")
fig.plot(timestamp[2000:28000], acc_x[2000:28000], label="UP Measured Acceleration")
fig.plot(sim_time, acc_up, color='black', lw=1.2, label="Simulated Acceleration")
fig.xlim([-1, 32])
fig.ylim([-40, 110])
fig.legend()
fig.show()
```

The burn was a little more progressive and shorter than in real life.

## Pressure Altitude

We also had a MPL3115A2 pressure sensor onboard. Looking at the raw pressure:

```python
columns = numpy.loadtxt("fcf/MPL3.csv", delimiter=',', unpack=True)

seqn = columns[0]
mpl_time = columns[1]
pressure = columns[2]

mpl_time = numpy.subtract(mpl_time, t_0)
mpl_time = numpy.divide(mpl_time, 1e9)

fig = utils.Plot("Launch 11 Measured Pressure", "Time [min:sec]", "Pressure [kPa]")
fig.plot(mpl_time, pressure, lw=1.2)
fig.timex()
fig.xlim([-5, 300])
fig.show()
```

Using the aproximation in the MPL3115A2 datasheet (based on a standard atmosphere) we can aproximate the altitude:


$$ h = 44330.77 \left[ 1 -  \left(\frac{p}{p_0}\right)^{0.1902632}\right]$$

```python
def alt(p):
    return 44330.77 * (1 - float(p/p_0)**0.1902632)

h = []
for p in pressure:
    h.append(alt(p))

# subtract launch altutude
alt_launch = (h[0]+h[1])/2.0
h = numpy.subtract(h, alt_launch)

# Max alt?
apogee = numpy.argmax(h)
apogee = (mpl_time[apogee], h[apogee])

fig = utils.Plot("Launch 11 Aprox Altitude", "Time [min:sec]", "Altitude(AGL) [m]")
fig.plot(mpl_time, h, lw=1.2)
fig.xlim([-5, 300])
fig.timex()
fig.note('Apogee (%0.0f m)' % apogee[1], (apogee[0], apogee[1]), (100,20))
fig.show()
```
