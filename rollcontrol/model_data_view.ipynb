# Roll Control Model Data

See `pipeline.py` for how the data used here was created. We import data from a previous flight:

```python
# Setup
import numpy as np
from bokeh.plotting import *
output_notebook()

# chart defaults
red = '#e31d1d'
blue = '#709afa'
linew = 1.4
w = 980
```

```python
# Load the data file and store in arrays
columns = np.loadtxt('./model_data.csv', delimiter=',', unpack=True)

time = columns[0]
angular_accel = columns[1]
angular_accel_filtered = columns[2]
velocity = columns[3]
altitude = columns[4]
fin_angle = columns[5]
roll_rate = columns[6]
```

---------------------

## Rocket Motion

First lets take a peak at the vertical motion of the rocket. Integrating the accelerometer once gets us a pretty good guess for velocity:

```python
line(time, velocity, title=u"Velocity", line_color=red, line_width=linew, plot_width=w)
yaxis().axis_label='velocity [m/s]'
xaxis().axis_label='Time [s]'
grid().grid_line_color='#eee'
show()
```

And integrating that gives us an altitude estimate too.

```python
line(time, np.divide(altitude, 1000.0), title=u"Altitude", line_color=red, line_width=linew, plot_width=w)
yaxis().axis_label='altitude [km]'
xaxis().axis_label='Time [s]'
grid().grid_line_color='#eee'
show()
```

---------------------------------

## Roll Data

Now let's look at the roll rate of the rocket.

```python
line(time, roll_rate, title=u"Roll rate/fin position", line_color=red, line_width=linew, plot_width=w, legend='roll rate')
hold()
line(time, np.multiply(fin_angle,10), line_color=blue, line_alpha=0.8, line_width=linew, plot_width=w, legend='fin position (exaggerated)')
yaxis().axis_label=u'[\u00ba/s]'
xaxis().axis_label='Time [s]'
grid().grid_line_color='#eee'
legend()[0].border_line_color = '#f7f7f7'
hold()
show()
```

---------------------

## Angular acceleration

The rocket has rate gyroscopes, so we only get direct measurements of roll rate, not angular acceleration. It's acceleration that is the result of an equation of motion, and is a key value in our control system. We get angular acceleration by differentiating the rate, but this is a very noisy process, so a low pass filter was applied as well.

```python
line(time, angular_accel, title=u"Angular acceleration", line_color=red, line_alpha=0.7, line_width=linew, plot_width=w, legend='unfiltered')
hold()
line(time, angular_accel_filtered, line_color=blue, line_width=linew, plot_width=w, legend='filtered')
yaxis().axis_label=u'[\u00ba/s/s]'
xaxis().axis_label='Time [s]'
grid().grid_line_color='#eee'
legend()[0].border_line_color = '#f7f7f7'
hold()
show()
```
