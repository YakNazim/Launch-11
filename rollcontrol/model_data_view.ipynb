See `pipeline.py` for how the data used here was created. We import data from a previous flight:

```python
#Imports
import numpy as np
from bokeh.plotting import *
output_notebook()
```

```python
# Load the data file and store in arrays
columns = np.loadtxt('./model_data.csv', delimiter=',', unpack=True)

time = columns[0]
angular_accel = columns[1]
velocity = columns[3]
fin_angle = columns[4]
```

## Fin Deflection vs Velocity

```python
# A plot
scatter(velocity, fin_angle, title=u"Fin Angle \u03b1 vs. Velocity")
xaxis().axis_label='velocity [m/s]' 
yaxis().axis_label=u'\u03b1'
grid().grid_line_color='#eee'

show()
```
